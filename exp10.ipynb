```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def gamma_correction(img, gamma):
    img_norm = img / 255.0
    return np.uint8((img_norm ** gamma) * 255)

gray = cv2.imread('/content/drive/MyDrive/Colab/assets/tigergrey.jpeg', 0)
color = cv2.imread('/content/drive/MyDrive/Colab/assets/tiger.jpeg')

gammas = [0.2, 0.5, 1.5, 2, 3]

for g in gammas:
    cv2_imshow(gamma_correction(gray, g))
    cv2_imshow(gamma_correction(color, g))
```
