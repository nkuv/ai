```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Colab/assets/tiger.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel_3x3 = np.ones((3, 3), np.float32) / 9
kernel_5x5 = np.ones((5, 5), np.float32) / 25
kernel_7x7 = np.ones((7, 7), np.float32) / 49

filtered_3x3 = cv2.filter2D(img_rgb, -1, kernel_3x3)
filtered_5x5 = cv2.filter2D(img_rgb, -1, kernel_5x5)
filtered_7x7 = cv2.filter2D(img_rgb, -1, kernel_7x7)

filtered_3x3_bgr = cv2.cvtColor(filtered_3x3, cv2.COLOR_RGB2BGR)
filtered_5x5_bgr = cv2.cvtColor(filtered_5x5, cv2.COLOR_RGB2BGR)
filtered_7x7_bgr = cv2.cvtColor(filtered_7x7, cv2.COLOR_RGB2BGR)

print("Original image")
cv2_imshow(img)

print("Filtered image with 3x3 kernel")
cv2_imshow(filtered_3x3_bgr)

print("Filtered image with 5x5 kernel")
cv2_imshow(filtered_5x5_bgr)

print("Filtered image with 7x7 kernel")
cv2_imshow(filtered_7x7_bgr)
```
