```python
import cv2
import os
from google.colab.patches import cv2_imshow

img = cv2.imread('assets/test_image.png')

for q in [90, 70, 50, 30, 10]:
    fname = f"img_{q}.jpeg"
    cv2.imwrite(fname, img, [cv2.IMWRITE_JPEG_QUALITY, q])

    fsize = os.path.getsize(fname)
    print(f"Quality = {q}, Size = {fsize} Bytes")

    img_new = cv2.imread(fname)
    cv2_imshow(img_new)
```
