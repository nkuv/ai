```python
import cv2
import os
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Colab/assets/tiger.jpg")
cv2_imshow(img)
```

```python
qualities = [90, 70, 50, 30, 10]

for q in qualities:
    fname = f"jpeg_{q}.jpeg"
    cv2.imwrite(fname, img, [cv2.IMWRITE_JPEG_QUALITY, q])

    size = os.path.getsize(fname) // 1024
    print(f"Quality={q} ,Filename={fname}, Size={size} KB")

    cv2_imshow(cv2.imread(fname))
```
