```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/Colab/assets/tiger.jpeg')

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

v_eq = cv2.equalizeHist(v)

hsv_eq = cv2.merge([h, s, v_eq])
img_eq = cv2.cvtColor(hsv_eq, cv2.COLOR_HSV2BGR)

cv2_imshow(img)
cv2_imshow(img_eq)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title('Histogram of V Before Equalization')
plt.hist(v.ravel(), bins=256, range=[0,256], color='blue')
plt.xlim([0,256])

plt.subplot(1,2,2)
plt.title('Histogram of V After Equalization')
plt.hist(v_eq.ravel(), bins=256, range=[0,256], color='blue')
plt.xlim([0,256])

plt.show()
```
