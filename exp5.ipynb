```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = np.zeros((200, 200), dtype=np.uint8)
img2 = np.zeros((200, 200), dtype=np.uint8)

cv2.rectangle(img1, (50, 50), (150, 150), 255, -1)
cv2.circle(img2, (100, 100), 60, 255, -1)
```

```python
bitwise_and = cv2.bitwise_and(img1, img2)
bitwise_or = cv2.bitwise_or(img1, img2)
bitwise_xor = cv2.bitwise_xor(img1, img2)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2, 3, 1); plt.imshow(img1, cmap='gray'); plt.title("Binary Image 1"); plt.axis('off')
plt.subplot(2, 3, 2); plt.imshow(img2, cmap='gray'); plt.title("Binary Image 2"); plt.axis('off')
plt.subplot(2, 3, 3); plt.imshow(bitwise_and, cmap='gray'); plt.title("AND"); plt.axis('off')
plt.subplot(2, 3, 4); plt.imshow(bitwise_or, cmap='gray'); plt.title("OR"); plt.axis('off')
plt.subplot(2, 3, 5); plt.imshow(bitwise_xor, cmap='gray'); plt.title("XOR"); plt.axis('off')
plt.tight_layout(); plt.show()
```
