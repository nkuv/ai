```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('assets/tiger.jpg')

gauss_3 = cv2.GaussianBlur(img, (3,3), 0)
gauss_5 = cv2.GaussianBlur(img, (5,5), 0)
gauss_7 = cv2.GaussianBlur(img, (7,7), 0)

print("Original image")
cv2_imshow(img)

print("Blurred image 3x3")
cv2_imshow(gauss_3)

print("Blurred image 5x5")
cv2_imshow(gauss_5)

print("Blurred image 7x7")
cv2_imshow(gauss_7)
```
