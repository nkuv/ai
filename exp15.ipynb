```python
import cv2
import os
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Colab/assets/tiger.jpg")

quality_levels = [100, 70, 50, 30, 10]
output_dir = "webp_images"
os.makedirs(output_dir, exist_ok=True)

for quality in quality_levels:
    output_path = os.path.join(output_dir, f"image_quality_{quality}.webp")
    cv2.imwrite(output_path, img, [cv2.IMWRITE_WEBP_QUALITY, quality])

    file_size_kb = os.path.getsize(output_path) / 1024

    img_quality = cv2.imread(output_path)

    print(f"Quality: {quality} - File size: {file_size_kb:.2f} KB")
    cv2_imshow(img_quality)
```
