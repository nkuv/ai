```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Colab/assets/tiger.jpg")

median_3 = cv2.medianBlur(img,3)
median_5 = cv2.medianBlur(img,5)
median_7 = cv2.medianBlur(img,7)

print("Original image")
cv2_imshow(img)

print("Blurred image 3x3")
cv2_imshow(median_3)

print("Blurred image 5x5")
cv2_imshow(median_5)

print("Blurred image 7x7")
cv2_imshow(median_7)
```
