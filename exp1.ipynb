```python
import cv2
from google.colab.patches import cv2_imshow
I = cv2.imread('/content/drive/MyDrive/Colab/assets/tiger.jpeg',-1)
cv2_imshow(I)
```

```python
print(I.shape)
```

```python
cv2.imwrite('/content/drive/MyDrive/Colab/assets/tiger2.jpeg',I)
```

```python
grey_image = cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)
cv2.imwrite('/content/drive/MyDrive/Colab/assets/tigergrey.jpeg', grey_image)
```

```python
cv2_imshow(grey_image)
```

```python
_, binary = cv2.threshold(grey_image,127,255,cv2.THRESH_BINARY)
cv2_imshow(binary)
cv2.imwrite('/content/drive/MyDrive/Colab/assets/tigerbinary.jpeg', binary)
```
