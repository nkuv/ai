```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/Colab/assets/tiger.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
tx, ty = 100, 50
rows, cols = img.shape[:2]
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated_img = cv2.warpAffine(img, translation_matrix, (cols, rows))

scaled_img = cv2.resize(img, None, fx=0.5, fy=0.5)

angle = 45
rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
rotated_img = cv2.warpAffine(img, rotation_matrix, (cols, rows))
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 4, 1); plt.imshow(img); plt.title("Original"); plt.axis('off')
plt.subplot(1, 4, 2); plt.imshow(translated_img); plt.title("Translated"); plt.axis('off')
plt.subplot(1, 4, 3); plt.imshow(scaled_img); plt.title("Scaled"); plt.axis('off')
plt.subplot(1, 4, 4); plt.imshow(rotated_img); plt.title("Rotated"); plt.axis('off')
plt.tight_layout(); plt.show()
```
