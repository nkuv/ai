```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/Colab/assets/tiger.jpeg')
cv2_imshow(img)
```

```python
blue = img[:, :, 0]
green = img[:, :, 1]
red = img[:, :, 2]

gray = (0.114 * blue + 0.587 * green + 0.299 * red)
gray8bit = gray.astype(np.uint8)

print("Grayscale image converted successfully.")
cv2_imshow(gray8bit)
```

```python
threshold = 127
mask = gray8bit > threshold
binary_img = np.zeros_like(gray8bit)
binary_img[mask] = 255

print("Binary image converted successfully.")
cv2_imshow(binary_img)
```
