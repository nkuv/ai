```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/drive/MyDrive/Colab/assets/tiger.jpeg')
img2 = cv2.imread('/content/drive/MyDrive/Colab/assets/cat.jpeg')
img1 = cv2.resize(img1, (400, 400))
img2 = cv2.resize(img2, (400, 400))
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
added = cv2.add(img1, img2)

subtracted = cv2.subtract(img1, img2)

blended = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1); plt.imshow(img1); plt.title("Image 1"); plt.axis('off')
plt.subplot(2, 2, 2); plt.imshow(img2); plt.title("Image 2"); plt.axis('off')
plt.subplot(2, 2, 3); plt.imshow(added); plt.title("Added"); plt.axis('off')
plt.subplot(2, 2, 4); plt.imshow(blended); plt.title("Blended"); plt.axis('off')
plt.tight_layout(); plt.show()
```
