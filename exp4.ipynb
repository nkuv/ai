```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img1 = cv2.imread('assets/tiger.jpeg')
img2 = cv2.imread('assets/cat.jpeg')

img1 = cv2.resize(img1, (400, 400))
img2 = cv2.resize(img2, (400, 400))

added_img = cv2.add(img1, img2)
subtracted_img = cv2.subtract(img1, img2)
blended_img = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

print("Displaying Image 1...")
cv2_imshow(img1)
print("Displaying Image 2...")
cv2_imshow(img2)
print("Displaying Added Image...")
cv2_imshow(added_img)
print("Displaying Subtracted Image...")
cv2_imshow(subtracted_img)
print("Displaying Blended Image...")
cv2_imshow(blended_img)
```
