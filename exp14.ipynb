```python
import cv2
import os
from google.colab.patches import cv2_imshow

img = cv2.imread('assets/test_image.png')

for c in [0, 3, 5, 7, 9]:
    fname = f"img_{c}.png"
    cv2.imwrite(fname, img, [cv2.IMWRITE_PNG_COMPRESSION, c])

    fsize = os.path.getsize(fname)
    print(f"Compression={c}, Size={fsize} Bytes")

    img_new = cv2.imread(fname)
    cv2_imshow(img_new)
```
