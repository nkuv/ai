```python
import cv2
import os
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/drive/MyDrive/Colab/assets/cat.jpeg")
cv2_imshow(img)
```

```python
compress_levels = [0, 3, 5, 7, 9]

for c in compress_levels:
    fname = f"png_{c}.png"
    cv2.imwrite(fname, img, [cv2.IMWRITE_PNG_COMPRESSION, c])

    size = os.path.getsize(fname) // 1024
    print(f"Compression={c}, Filename={fname}, Size={size} KB")

    # Display compressed image
    cv2_imshow(cv2.imread(fname))
```
